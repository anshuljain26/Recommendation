```python
import pandas as pd
#movies file
movies = pd.read_csv('ml-latest-small/movies.csv', usecols=range(2))
movies.head()
```

```python
# Ratings file
ratings = pd.read_csv('ml-latest-small/ratings.csv',usecols=range(3))
ratings.head()
```

```python
ratings = pd.merge(movies,ratings)
ratings.head()
```

```python
movie_ratings = ratings.pivot_table(index=['userId'],columns=['title'], values='rating')
movie_ratings.head()
# A table containg user rating corresponding to every movies
```

```python
# find out similar movie corresponding to Assassins movie
rating = movie_ratings['Assassins (1995)']
similar_movies = movie_ratings.corrwith(rating)
similar_movies = similar_movies.dropna()
df  = pd.DataFrame(similar_movies)
df.head()
```

```python
similar_movies.order(ascending=False)
```

```python
import numpy as np
moviestats = ratings.groupby('title').agg({'rating':[np.size,np.mean]})
moviestats.head()
```

```python
# find out the popular_movies rated by more than 100 users
popular_movies = moviestats['rating']['size']>=100
moviestats[popular_movies].sort_values([('rating','mean')],ascending=False)[:10]
```

```python
df = moviestats[popular_movies].join(pd.DataFrame(similar_movies,columns=['similarity']))
df.sort_values(['similarity'],ascending=False)[:10]
```

```python
# Recommend the movie to a random user by using item based collaborate filtering algorithm
movie_ratings.head()
```

```python
user = movie_ratings.iloc[50].dropna()
data = pd.Series()
for i in range(len(user)):
    rating = movie_ratings[user.index[i]]
    similar_movies = movie_ratings.corrwith(rating).dropna()
    #print similar_movies corresponding to each movie
    similar_movies = similar_movies.map(lambda x: x*user[i])
    # append the movies into data dataframe
    data = data.append(similar_movies)

data.sort_values(inplace=True,ascending=False)
#data.head(5)
#print len(data)
```

```python
data = data.groupby(data.index).sum()
data.sort_values(inplace=True,ascending=False)
data.head(5)
```

```python
recommend_movies = moviestats[popular_movies].join(pd.DataFrame(data,columns=['similarity']))
#recommend_movies.sort_values(['similarity'], ascending=False)
# here is the final recommended movies to a user
recommend_movies.sort_values(['similarity'],ascending=False)[:10]
#print len(recommend_movies)
```
